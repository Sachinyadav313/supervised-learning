```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
dataSet= pd.read_csv("Housing.csv")
```

```python
#defining the values of x and y which is our feature and target value
x=dataSet.price
y=dataSet.area
x=np.array(x)
y=np.array(y)
```

```python
# Normalize the data
x = (x - np.min(x)) / (np.max(x) - np.min(x))
y=(y - np.min(y)) / (np.max(y) - np.min(y))
```

```python
plt.scatter(x, y, marker='x',c='r')  # 'x' for dots at data points
plt.title('Linear Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
  # Add grid
plt.show()
```

```python
class LinearModel:
    def __init__(self, w, b):
        self.w = w
        self.b = b

    def predict(self, x):
        output = np.zeros_like(x)
        for i in range(len(x)):
            output[i] = self.w * x[i] + self.b
        return output
# Create an instance of the LinearModel class
model = LinearModel(w=.2, b=.1)

# Use the model to make predictions
output = model.predict(x)

# Plotting
plt.plot(x, output, c='b', label='Our Prediction')
plt.scatter(x, y, marker='x', c='r', label='Actual Values')
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
