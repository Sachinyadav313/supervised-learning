```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv("Housing.csv")
print(dataset.head())
```

```python
x=dataset.price
y=dataset.area
x=np.array(x)
y=np.array(y)
```

```python
x = (x - np.min(x)) / (np.max(x) - np.min(x))
y=(y - np.min(y)) / (np.max(y) - np.min(y))
```

```python
 plt.scatter(x, y, marker='x',c='r')  # 'x' for dots at data points
plt.title('Linear Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
  # Add grid
plt.show()
```

```python
class CostFunction:
    def __init__(self, w, b):
        self.w = w
        self.b = b
    
    def cost(self, x, y):
        size = len(x)
        total = 0
        for i in range(size):
            y_pred = self.w * x[i] + self.b
            value_cost = (y_pred - y[i]) ** 2
            total += value_cost
        total_sum = (1 / (2 * size)) * total
        return total_sum
model = CostFunction(w=1, b=1)
result = model.cost(x, y)
print(result)
```
